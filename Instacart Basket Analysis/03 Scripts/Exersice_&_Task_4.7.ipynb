```python
# Create the subset

df = df_ords_prods_merge[:1000000]
```

```python
# If-Statements with User-Defined Functions


def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df.shape
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# If-Statements with the loc() Function

df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Create a new column for "Busiest day" using a For-Loops method.

result_n2 = []

for value in df_ords_prods_merge["order_dow"]:
  if value == 0 or value == 1:
    result_n2.append("Busiest days")
  elif value == 4 or value == 3:
    result_n2.append("Least busy days")
  else:
    result_n2.append("Regularly busy")
```

```python
result_n2
```

```python
df_ords_prods_merge["Busiest days"] = result_n2
```

```python
# Check the values of this new column "Busiest days"

df_ords_prods_merge["Busiest days"].value_counts(dropna = False)
```

```python
# Create a new column containing Most orders, Average orders, and Fewest orders labels called busiest_period_of_day


df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Group data into "Most orders," "Average orders," and "Fewest orders" labels and use the For-Loops method

result_period_of_day = []

for value in df_ords_prods_merge['order_hour_of_day']:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_period_of_day.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_period_of_day.append("Fewest orders")
  else:
    result_period_of_day.append("Average orders")
```

```python
# Adding the result to df as "busiest_period_of_day" column

df_ords_prods_merge["busiest_period_of_day"] = result_period_of_day
```
